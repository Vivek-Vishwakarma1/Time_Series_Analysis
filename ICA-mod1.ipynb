```python
import pandas as pd
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
np.random.seed(23) #setting up seed for random number generator for reproducibility
```

```python
ns = np.linspace(0, 200, 1000) #no of samples

#Source matrix

S = np.array([np.sin(ns * 1), 
              signal.sawtooth(ns*1.9),
              np.random.random(len(ns))]).T

#mixing matrix
A = np.array([[0.5, 1, 0.2],
             [1, 0.5, 0.4],
             [0.5, 0.8, 1]])

#Mixing signal matrix
X = S.dot(A).T
```

```python
fig, ax = plt.subplots(1, 1, figsize = [18, 5])
ax.plot(ns, S, lw = 1)
ax.set_xticks([])
ax.set_yticks([-1,1])
ax.set_xlim(ns[0], ns[200])
ax.tick_params(labelsize = 12)
ax.set_title('Independent Compoenet Analysis', fontsize = 20)
```

```python
fig, ax = plt.subplots(3, 1, figsize = [18, 5], sharex = True)
ax[0].plot(ns, X[0], lw=1)
ax[0].set_title('Mixed signals', fontsize = 25)
ax[0].tick_params(labelsize=12)

ax[1].plot(ns, X[1], lw=1)
ax[1].tick_params(labelsize = 12)
ax[1].set_xlim(ns[0], ns[-1])

ax[2].plot(ns, X[2], lw=1)
ax[2].tick_params(labelsize = 12)
ax[2].set_xlim(ns[0], ns[-1])
ax[2].set_xlabel('Sample number', fontsize = 20)
ax[2].set_xlim(ns[0], ns[200])
```

```python
#Defining two non-gaussian components
s1 = np.random.rand(1000)
s2 = np.random.rand(1000)
s = np.array([s1, s2])
```

```python
#Defining two gaussian components
s1n = np.random.normal(size=1000)
s2n = np.random.normal(size = 1000)
sn = np.array(([s1n, s2n]))
```

```python
#Defining orthogonal matrix
A = np.array(([0.96, -0.28], [0.28, 0.96]))
```

```python
#Mix signal
mixsignals = s.T.dot(A)
mixsignalsN = sn.T.dot(A)
```

```python
fig, ax = plt.subplots(2,2,figsize = [18, 10])
ax[0][0].scatter(s[0], s[1])
ax[0][0].tick_params(labelsize = 10)
ax[0][0].set_title('Sources (non-Gausian)', fontsize = 25)
ax[0][0].set_xlim([-0.25, 1.5])
ax[0][0].set_xticks([])

ax[0][1].scatter(sn[0], sn[1])
ax[0][1].tick_params(labelsize = 12)
ax[0][1].set_title('Source (Gaussian)', fontsize = 25)
ax[0][1].set_xlim(-4, 4)
ax[0][1].set_xticks([])
ax[0][1].set_yticks([])

ax[1][0].scatter(mixsignals.T[0], mixsignals.T[1])
ax[1][0].tick_params(labelsize = 12)
ax[1][0].set_title('Mixed signals (non-Gausaian sources)', fontsize = 25)
ax[1][0].set_xlim([-0.25, 1.5])

ax[1][1].scatter(mixsignalsN.T[0], mixsignalsN.T[1])
ax[1][1].tick_params(labelsize = 12)
ax[1][1].set_xlim([-4, 4])
ax[1][1].set_title('Mixed signal (Gaussian source)', fontsize = 25)
ax[1][1].set_yticks([])
```
